Resume screening usinh AI and Machine Learning

```python
def analyze_resume(resume_text):
    """
    Analyzes a resume text and determines if the candidate should be selected.

    Args:
        resume_text (str): The resume content in text format

    Returns:
        bool: True if selected, False otherwise
        dict: Analysis details including score and matched keywords
    """
    # Convert resume to lowercase for case-insensitive matching
    resume_text = resume_text.lower()

    # Define keywords to look for (with weights)
    keywords = {
        # Technical skills
        'python': 5,
        'java': 4,
        'javascript': 4,
        'html': 2,
        'css': 2,
        'sql': 4,
        'machine learning': 5,
        'data analysis': 4,
        'aws': 4,
        'docker': 3,
        'kubernetes': 3,
        'git': 3,

        # Education
        'phd': 5,
        'master': 4,
        'bachelor': 3,
        'computer science': 4,
        'engineering': 3,

        # Experience indicators
        'years of experience': 3,
        'project manager': 4,
        'team lead': 4,
        'developer': 3,
        'engineer': 3,

        # Soft skills
        'leadership': 3,
        'communication': 3,
        'teamwork': 3,
        'problem solving': 4
    }

    # Calculate score based on keyword matches
    score = 0
    matched_keywords = {}

    for keyword, weight in keywords.items():
        if keyword in resume_text:
            score += weight
            matched_keywords[keyword] = weight

    # Define threshold for selection
    threshold = 20

    # Determine if candidate should be selected
    selected = score >= threshold

    # Prepare analysis details
    analysis = {
        'score': score,
        'threshold': threshold,
        'matched_keywords': matched_keywords,
        'selection_status': 'Selected' if selected else 'Not Selected'
    }

    return selected, analysis


def main():
    print("Resume Analyzer - Enter resume text (type 'END' on a new line when finished):")

    # Collect resume text from user
    resume_lines = []
    while True:
        line = input()
        if line.strip().upper() == 'END':
            break
        resume_lines.append(line)

    resume_text = '\n'.join(resume_lines)

    # Analyze the resume
    selected, analysis = analyze_resume(resume_text)

    # Display results
    print("\n" + "="*50)
    print("RESUME ANALYSIS RESULTS")
    print("="*50)
    print(f"Selection Status: {analysis['selection_status']}")
    print(f"Score: {analysis['score']} (Threshold: {analysis['threshold']})")

    print("\nMatched Keywords:")
    if analysis['matched_keywords']:
        for keyword, weight in sorted(analysis['matched_keywords'].items(), key=lambda x: x[1], reverse=True):
            print(f"- {keyword} (weight: {weight})")
    else:
        print("- No keywords matched")

    print("\nRecommendation:")
    if selected:
        print("This candidate meets the minimum requirements and should be considered for the next round.")
    else:
        print("This candidate does not meet the minimum requirements based on keyword analysis.")
        if analysis['score'] >= analysis['threshold'] * 0.7:
            print("However, the score is close to the threshold. Consider reviewing manually.")


if __name__ == "__main__":
    main()
```

```python
def get_company_requirements():
    """Collect company requirements from the user"""
    print("\n=== COMPANY REQUIREMENTS ===")

    job_title = input("Enter job title: ")

    print("\nEnter required skills (comma-separated):")
    skills_input = input()
    required_skills = [skill.strip().lower() for skill in skills_input.split(',')]

    print("\nEnter minimum education level (e.g., Bachelor's, Master's, PhD):")
    min_education = input().lower()

    print("\nEnter minimum years of experience required:")
    min_experience = int(input())

    print("\nEnter any additional keywords to look for (comma-separated):")
    keywords_input = input()
    additional_keywords = [keyword.strip().lower() for keyword in keywords_input.split(',')]

    return {
        "job_title": job_title,
        "required_skills": required_skills,
        "min_education": min_education,
        "min_experience": min_experience,
        "additional_keywords": additional_keywords
    }

def extract_years_of_experience(resume_text):
    """Extract years of experience from resume text"""
    import re

    # Look for patterns like "X years of experience" or "X+ years"
    patterns = [
        r'(\d+)\+?\s*(?:years|yrs)(?:\s*of)?\s*experience',
        r'experience\s*(?:of|for)?\s*(\d+)\+?\s*(?:years|yrs)',
        r'worked\s*(?:for)?\s*(\d+)\+?\s*(?:years|yrs)'
    ]

    max_years = 3
    for pattern in patterns:
        matches = re.findall(pattern, resume_text.lower())
        for match in matches:
            years = int(match)
            max_years = max(max_years, years)

    return max_years

def detect_education_level(resume_text):
    """Detect highest education level from resume text"""
    education_levels = {
        'phd': 4,
        'doctorate': 4,
        'doctor of': 4,
        'master': 3,
        'mba': 3,
        'ms ': 3,
        'ma ': 3,
        'bachelor': 2,
        'bs ': 2,
        'ba ': 2,
        'associate': 1,
        'high school': 0,
        'diploma': 0
    }

    resume_lower = resume_text.lower()
    highest_level = 0
    highest_degree = "None detected"

    for degree, level in education_levels.items():
        if degree in resume_lower:
            if level > highest_level:
                highest_level = level
                highest_degree = degree

    education_mapping = {
        0: "High School",
        1: "Associate's",
        2: "Bachelor's",
        3: "Master's",
        4: "PhD"
    }

    return education_mapping.get(highest_level, "Unknown"), highest_level

def analyze_resume_against_requirements(resume_text, requirements):
    """
    Analyze resume against company requirements

    Args:
        resume_text (str): The resume content
        requirements (dict): Company requirements

    Returns:
        dict: Analysis results
    """
    resume_lower = resume_text.lower()

    # Check for required skills
    found_skills = []
    missing_skills = []
    for skill in requirements["required_skills"]:
        if skill in resume_lower:
            found_skills.append(skill)
        else:
            missing_skills.append(skill)

    skills_match_percentage = 0
    if requirements["required_skills"]:
        skills_match_percentage = (len(found_skills) / len(requirements["required_skills"])) * 100

    # Check education level
    detected_education, education_level = detect_education_level(resume_text)

    education_levels_map = {
        "high school": 0,
        "associate's": 1,
        "bachelor's": 2,
        "master's": 3,
        "phd": 4
    }

    required_education_level = education_levels_map.get(requirements["min_education"].lower(), 0)
    education_meets_requirement = education_level >= required_education_level

    # Check years of experience
    detected_experience = extract_years_of_experience(resume_text)
    experience_meets_requirement = detected_experience >= requirements["min_experience"]

    # Check for additional keywords
    found_keywords = []
    for keyword in requirements["additional_keywords"]:
        if keyword and keyword in resume_lower:
            found_keywords.append(keyword)

    # Calculate overall match percentage
    criteria_count = 3  # Skills, education, experience
    criteria_met = 0

    if skills_match_percentage >= 70:
        criteria_met += 1
    if education_meets_requirement:
        criteria_met += 1
    if experience_meets_requirement:
        criteria_met += 1

    overall_match = (criteria_met / criteria_count) * 100

    # Determine if candidate should be selected
    selected = overall_match >= 70

    return {
        "selected": selected,
        "overall_match_percentage": overall_match,
        "skills_match_percentage": skills_match_percentage,
        "found_skills": found_skills,
        "missing_skills": missing_skills,
        "detected_education": detected_education,
        "education_meets_requirement": education_meets_requirement,
        "detected_experience": detected_experience,
        "experience_meets_requirement": experience_meets_requirement,
        "found_keywords": found_keywords
    }

def main():
    # Get company requirements
    requirements = get_company_requirements()

    print("\n=== RESUME INPUT ===")
    print("Enter resume text (type 'END' on a new line when finished):")

    # Collect resume text from user
    resume_lines = []
    while True:
        line = input()
        if line == 'END':
            break
        resume_lines.append(line)

    resume_text = '\n'.join(resume_lines)

    # Analyze the resume against requirements
    results = analyze_resume_against_requirements(resume_text, requirements)

    # Display results
    print("\n" + "="*60)
    print(f"RESUME ANALYSIS RESULTS FOR: {requirements['job_title']}")
    print("="*60)

    print(f"\nSELECTION STATUS: {'SELECTED' if results['selected'] else 'NOT SELECTED'}")
    print(f"Overall Match: {results['overall_match_percentage']:.1f}%")

    print("\n--- SKILLS ANALYSIS ---")
    print(f"Skills Match: {results['skills_match_percentage']:.1f}%")
    print("Found Skills:")
    for skill in results['found_skills']:
        print(f"  ✓ {skill}")
    print("Missing Skills:")
    for skill in results['missing_skills']:
        print(f"  ✗ {skill}")

    print("\n--- EDUCATION ANALYSIS ---")
    print(f"Detected Education: {results['detected_education']}")
    print(f"Required Education: {requirements['min_education']}")
    print(f"Requirement Met: {'Yes' if results['education_meets_requirement'] else 'No'}")

    print("\n--- EXPERIENCE ANALYSIS ---")
    print(f"Detected Experience: {results['detected_experience']} years")
    print(f"Required Experience: {requirements['min_experience']} years")
    print(f"Requirement Met: {'Yes' if results['experience_meets_requirement'] else 'No'}")

    if results['found_keywords']:
        print("\n--- ADDITIONAL KEYWORDS FOUND ---")
        for keyword in results['found_keywords']:
            print(f"  • {keyword}")

    print("\n--- RECOMMENDATION ---")
    if results['selected']:
        print("This candidate meets the minimum requirements for the position.")
        if results['overall_match_percentage'] >= 90:
            print("Strong match! Consider prioritizing this candidate for interviews.")
        else:
            print("Candidate meets requirements but may need additional screening.")
    else:
        print("This candidate does not meet the minimum requirements.")
        if results['overall_match_percentage'] >= 50:
            print("However, the candidate shows potential. Consider for a different role or keep on file.")
        else:
            print("Not recommended for further consideration at this time.")

if __name__ == "__main__":
    main()
```
